```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.tri as mtri
```

```python
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
X,Y = np.meshgrid(x,y)
Z= 4*X + 8*Y + 1

X1,Y1 = np.meshgrid(x,y)
Z1= 4*X1 + 8*Y1 + 10

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X1, Y1, Z1)
plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 5 * np.outer(np.sin(u), np.sin(v))
z = 3 * np.outer(np.ones(np.size(u)), np.cos(v))

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()

n_radii = 8
n_angles = 36

radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)

angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)

x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())

z = np.sin(-x*y)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_trisurf(x, y, z, linewidth=0.2, antialiased=True, cmap=plt.cm.Spectral)

plt.show()
```

#### Решение:
Преобразование A n-мерного евклидова пространства называется ортогональным, если оно сохраняет скалярное произведение векторов. Из определения следуют простейшие свойства: при ортогональном преобразовании евклидова пространства не изменяются длины векторов, а также углы между векторами.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def polar(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y
```

```python
t = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)

plt.polar(x, y)
plt.show()
```

#### 4. Задание (в программе)
Решите систему уравнений:
$$ y = x^2 – 1 $$
$$ exp(x) + x \bullet (1 – y) = 1 $$
Решите систему уравнений и неравенств:
$$ y = x^2 – 1 $$
$$ exp(x) + x \bullet (1 – y) > 1 $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-6, 6, 200)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (((np.exp(x)) - 1) / x) + 1
plt.plot(x, y, label=f'$exp(x) + x∙(1 – y) = 1$')

plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-3, 5, 200)

y = np.exp(x) - 1 + x*(2-x**2)
plt.plot(x, y)
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def equations(x):
    return np.exp(x) - 1 + x*(2-x**2)
```

```python
print(f' x_1:{fsolve(equations, -1.5)[0]}\t y_1: {(fsolve(equations, -1.5)[0])**2 - 1}\n',\
      f'x_2:{fsolve(equations, 0)[0]}\t\t y_2: {(fsolve(equations, 0)[0])**2 - 1}\n',\
      f'x_3:{fsolve(equations, 2.5)[0]}\t y_3: {(fsolve(equations, 2.5)[0])**2 - 1}\n',\
      f'x_4:{fsolve(equations, 4)[0]}\t y_4: {(fsolve(equations, 4)[0])**2 - 1}'
     )
```
